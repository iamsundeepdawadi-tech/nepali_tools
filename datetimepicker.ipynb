```python
import ipywidgets as widgets
from IPython.display import display
import datetime

class NepaliDatePicker:
    """
    Jupyter Notebook को लागि ipywidgets प्रयोग गरी बनाइएको
    अन्तरक्रियात्मक नेपाली मिति पात्रो (Nepali Date Picker)।
    """

    def __init__(self, value=None):
        # JavaScript कोडबाट निकालिएको वि.सं. २००० देखि २१०० सम्मको पात्रो डाटा
        self.bs_calendar_data = [
            {'year': 2000, 'days': [30, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]}, {'year': 2001, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2002, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]}, {'year': 2003, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]},
            {'year': 2004, 'days': [30, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]}, {'year': 2005, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2006, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]}, {'year': 2007, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]},
            {'year': 2008, 'days': [31, 31, 31, 32, 31, 31, 29, 30, 30, 29, 29, 31]}, {'year': 2009, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2010, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]}, {'year': 2011, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]},
            {'year': 2012, 'days': [31, 31, 31, 32, 31, 31, 29, 30, 30, 29, 30, 30]}, {'year': 2013, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2014, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]}, {'year': 2015, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]},
            {'year': 2016, 'days': [31, 31, 31, 32, 31, 31, 29, 30, 30, 29, 30, 30]}, {'year': 2017, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2018, 'days': [31, 32, 31, 32, 31, 30, 30, 29, 30, 29, 30, 30]}, {'year': 2019, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]},
            {'year': 2020, 'days': [31, 31, 31, 32, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2021, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2022, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 30]}, {'year': 2023, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]},
            {'year': 2024, 'days': [31, 31, 31, 32, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2025, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2026, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]}, {'year': 2027, 'days': [30, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]},
            {'year': 2028, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2029, 'days': [31, 31, 32, 31, 32, 30, 30, 29, 30, 29, 30, 30]},
            {'year': 2030, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]}, {'year': 2031, 'days': [30, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]},
            {'year': 2032, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2033, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]},
            {'year': 2034, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]}, {'year': 2035, 'days': [30, 32, 31, 32, 31, 31, 29, 30, 30, 29, 29, 31]},
            {'year': 2036, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2037, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]},
            {'year': 2038, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]}, {'year': 2039, 'days': [31, 31, 31, 32, 31, 31, 29, 30, 30, 29, 30, 30]},
            {'year': 2040, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2041, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]},
            {'year': 2042, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]}, {'year': 2043, 'days': [31, 31, 31, 32, 31, 31, 29, 30, 30, 29, 30, 30]},
            {'year': 2044, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2045, 'days': [31, 32, 31, 32, 31, 30, 30, 29, 30, 29, 30, 30]},
            {'year': 2046, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]}, {'year': 2047, 'days': [31, 31, 31, 32, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2048, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2049, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 30]},
            {'year': 2050, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]}, {'year': 2051, 'days': [31, 31, 31, 32, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2052, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2053, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 30]},
            {'year': 2054, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]}, {'year': 2055, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2056, 'days': [31, 31, 32, 31, 32, 30, 30, 29, 30, 29, 30, 30]}, {'year': 2057, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]},
            {'year': 2058, 'days': [30, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]}, {'year': 2059, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2060, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]}, {'year': 2061, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]},
            {'year': 2062, 'days': [30, 32, 31, 32, 31, 31, 29, 30, 29, 30, 29, 31]}, {'year': 2063, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2064, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]}, {'year': 2065, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]},
            {'year': 2066, 'days': [31, 31, 31, 32, 31, 31, 29, 30, 30, 29, 29, 31]}, {'year': 2067, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2068, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]}, {'year': 2069, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]},
            {'year': 2070, 'days': [31, 31, 31, 32, 31, 31, 29, 30, 30, 29, 30, 30]}, {'year': 2071, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2072, 'days': [31, 32, 31, 32, 31, 30, 30, 29, 30, 29, 30, 30]}, {'year': 2073, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]},
            {'year': 2074, 'days': [31, 31, 31, 32, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2075, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2076, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 30]}, {'year': 2077, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]},
            {'year': 2078, 'days': [31, 31, 31, 32, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2079, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2080, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 30]}, {'year': 2081, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]},
            {'year': 2082, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2083, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]},
            {'year': 2084, 'days': [31, 31, 32, 31, 31, 30, 30, 30, 29, 30, 30, 30]}, {'year': 2085, 'days': [30, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]},
            {'year': 2086, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2087, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]},
            {'year': 2088, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]}, {'year': 2089, 'days': [30, 32, 31, 32, 31, 30, 30, 30, 29, 30, 29, 31]},
            {'year': 2090, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2091, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]},
            {'year': 2092, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]}, {'year': 2093, 'days': [31, 31, 31, 32, 31, 31, 29, 30, 30, 29, 29, 31]},
            {'year': 2094, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2095, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]},
            {'year': 2096, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]}, {'year': 2097, 'days': [31, 31, 31, 32, 31, 31, 29, 30, 30, 29, 30, 30]},
            {'year': 2098, 'days': [31, 31, 32, 31, 31, 31, 30, 29, 30, 29, 30, 30]}, {'year': 2099, 'days': [31, 31, 32, 32, 31, 30, 30, 29, 30, 29, 30, 30]},
            {'year': 2100, 'days': [31, 32, 31, 32, 31, 30, 30, 30, 29, 29, 30, 31]}
        ]
        
        # सन्दर्भ मिति: वि.सं. २००० बैशाख १ = सन् १९४३ अप्रिल १४
        self.ref_ad_date = datetime.date(1943, 4, 14)
        
        self.nepali_months = ["बैशाख", "जेठ", "असार", "श्रावण", "भाद्र", "आश्विन", "कार्तिक", "मंसिर", "पौष", "माघ", "फाल्गुण", "चैत्र"]
        self.week_days = ["आइत", "सोम", "मंगल", "बुध", "बिहि", "शुक्र", "शनि"]
        
        self.year_map = {y['year']: y for y in self.bs_calendar_data}
        
        # सुरुको मिति सेट गर्ने
        if value and isinstance(value, str):
            try:
                y, m, d = map(int, value.split('-'))
                self.selected_date = {'year': y, 'month': m, 'day': d}
            except:
                self.selected_date = self.ad_to_bs(datetime.date.today())
        else:
            self.selected_date = self.ad_to_bs(datetime.date.today())
        
        # Widget हरू बनाउने
        self._init_widgets()
        self._attach_observers()
        self.update_calendar()

    def ad_to_bs(self, ad_date):
        """अंग्रेजी मितिलाई नेपालीमा रूपान्तरण गर्छ"""
        day_diff = (ad_date - self.ref_ad_date).days
        
        total_days = 0
        bs_year, bs_month, bs_day = 0, 0, 0
        
        for year_data in self.bs_calendar_data:
            year_total_days = sum(year_data['days'])
            if total_days + year_total_days > day_diff:
                bs_year = year_data['year']
                day_diff_in_year = day_diff - total_days
                
                month_days = 0
                for month_idx, days_in_month in enumerate(year_data['days']):
                    if month_days + days_in_month > day_diff_in_year:
                        bs_month = month_idx + 1
                        bs_day = day_diff_in_year - month_days + 1
                        break
                    month_days += days_in_month
                break
            total_days += year_total_days
            
        return {'year': bs_year, 'month': bs_month, 'day': bs_day}

    def _init_widgets(self):
        """पात्रोको लागि आवश्यक सबै ipywidgets बनाउँछ"""
        years = list(self.year_map.keys())
        months = list(range(1, 13))

        self.year_dropdown = widgets.Dropdown(options=years, value=self.selected_date['year'])
        self.month_dropdown = widgets.Dropdown(options=[(self.nepali_months[m-1], m) for m in months], value=self.selected_date['month'])
        
        self.prev_year_btn = widgets.Button(description="<<", layout=widgets.Layout(width='50px'))
        self.next_year_btn = widgets.Button(description=">>", layout=widgets.Layout(width='50px'))
        self.prev_month_btn = widgets.Button(description="<", layout=widgets.Layout(width='50px'))
        self.next_month_btn = widgets.Button(description=">", layout=widgets.Layout(width='50px'))
        
        self.selected_date_label = widgets.Label(value=f"{self.selected_date['year']}-{self.selected_date['month']:02d}-{self.selected_date['day']:02d}")
        
        header_box = widgets.HBox([self.prev_year_btn, self.prev_month_btn, self.year_dropdown, self.month_dropdown, self.next_month_btn, self.next_year_btn])
        
        self.calendar_output = widgets.Output()
        
        self.container = widgets.VBox([
            widgets.HBox([widgets.Label("छानिएको मिति:"), self.selected_date_label]),
            header_box,
            self.calendar_output
        ])

    def _attach_observers(self):
        """Widget हरूमा event handler हरू (observers) जोड्छ"""
        self.year_dropdown.observe(self.handle_nav_change, names='value')
        self.month_dropdown.observe(self.handle_nav_change, names='value')
        
        self.prev_year_btn.on_click(self.handle_prev_year)
        self.next_year_btn.on_click(self.handle_next_year)
        self.prev_month_btn.on_click(self.handle_prev_month)
        self.next_month_btn.on_click(self.handle_next_month)

    def update_calendar(self):
        """पात्रोको दिनहरू भएको ग्रिडलाई पुनः कोर्छ (redraws)"""
        year = self.year_dropdown.value
        month = self.month_dropdown.value
        
        # महिनाको पहिलो गते कुन बार पर्छ पत्ता लगाउने
        total_days = 0
        for y_data in self.bs_calendar_data:
            if y_data['year'] < year:
                total_days += sum(y_data['days'])
            elif y_data['year'] == year:
                total_days += sum(y_data['days'][:month-1])
                break
        
        first_day_ad = self.ref_ad_date + datetime.timedelta(days=total_days)
        start_day_of_week = (first_day_ad.weekday() + 1) % 7 # 0=आइतबार, 1=सोमबार...
        
        days_in_month = self.year_map[year]['days'][month-1]
        
        # दिनहरू र बारको हेडर बनाउने
        week_headers = [widgets.Button(description=d, disabled=True, layout=widgets.Layout(width='50px')) for d in self.week_days]
        day_buttons = []
        
        # महिना सुरु हुनुअघिको खाली ठाउँ
        for _ in range(start_day_of_week):
            day_buttons.append(widgets.Label(""))
            
        # महिनाको दिनहरू
        for day in range(1, days_in_month + 1):
            btn = widgets.Button(description=str(day), layout=widgets.Layout(width='50px'))
            if year == self.selected_date['year'] and month == self.selected_date['month'] and day == self.selected_date['day']:
                btn.button_style = 'info' # छानिएको दिनलाई हाइलाइट गर्ने
            
            btn.on_click(self.handle_day_click)
            day_buttons.append(btn)
        
        grid = widgets.GridBox(week_headers + day_buttons, layout=widgets.Layout(grid_template_columns="repeat(7, 50px)"))
        
        with self.calendar_output:
            self.calendar_output.clear_output(wait=True)
            display(grid)
            
    # --- Event Handlers ---
    def handle_day_click(self, button):
        day = int(button.description)
        self.selected_date = {
            'year': self.year_dropdown.value,
            'month': self.month_dropdown.value,
            'day': day
        }
        self.selected_date_label.value = f"{self.selected_date['year']}-{self.selected_date['month']:02d}-{self.selected_date['day']:02d}"
        self.update_calendar()
        
    def handle_nav_change(self, change):
        self.update_calendar()

    def handle_prev_year(self, b):
        current_index = self.year_dropdown.options.index(self.year_dropdown.value)
        if current_index > 0:
            self.year_dropdown.value = self.year_dropdown.options[current_index - 1]

    def handle_next_year(self, b):
        current_index = self.year_dropdown.options.index(self.year_dropdown.value)
        if current_index < len(self.year_dropdown.options) - 1:
            self.year_dropdown.value = self.year_dropdown.options[current_index + 1]

    def handle_prev_month(self, b):
        current_month = self.month_dropdown.value
        if current_month > 1:
            self.month_dropdown.value = current_month - 1
        else:
            self.handle_prev_year(b)
            self.month_dropdown.value = 12

    def handle_next_month(self, b):
        current_month = self.month_dropdown.value
        if current_month < 12:
            self.month_dropdown.value = current_month + 1
        else:
            self.handle_next_year(b)
            self.month_dropdown.value = 1
            
    @property
    def value(self):
        """छानिएको मितिलाई 'YYYY-MM-DD' ढाँचामा string को रूपमा फर्काउँछ।"""
        return self.selected_date_label.value
        
    def _ipython_display_(self):
        """Jupyter मा widget प्रदर्शन गर्नको लागि"""
        display(self.container)


# --- Jupyter Notebook मा प्रयोगको लागि उदाहरण ---

# उदाहरण १: आजको मितिसहित पात्रो बनाउने
print("--- आजको मितिसहितको पात्रो ---")
datepicker1 = NepaliDatePicker()
display(datepicker1)

# केही समयपछि, छानिएको मिति हेर्नको लागि, तलको line अर्को सेलमा चलाउनुहोस्:
# print(f"तपाईंले छान्नुभएको मिति: {datepicker1.value}")

# उदाहरण २: सुरुमै कुनै निश्चित मिति दिएर पात्रो बनाउने
print("\n--- पूर्वनिर्धारित मिति '२०८०-०५-१०' सहितको पात्रो ---")
datepicker2 = NepaliDatePicker(value="2080-05-10")
display(datepicker2)
```
