```python
# आवश्यक लाइब्रेरी import गर्ने
import random

# --- १. नेपाली शब्दहरूको सूची (Nepali Word List) ---
nepali_words = [
    "राम्रो", "सुन्दर", "विश्व", "प्रकृति", "मन", "जीवन", "कथा", "भाषा", "विज्ञान", "कला",
    "इतिहास", "समाज", "साहित्य", "शिक्षा", "संस्कृति", "नेपाल", "परम्परा", "समाज",
    "प्रेम", "संगीत", "कलाकार", "धार्मिक", "सन्देश", "गाउँ", "शहर", "प्रयास", "सफलता", "माया", "सपना",
    "बुद्ध", "ज्ञान", "सकारात्मक", "धैर्य", "समर्पण", "अनुभव", "सत्य", "मूल्य", "शक्ति", "समय",
    "मनोबल", "साहस", "सन्तुष्टि", "गुणस्तर", "समाधान", "समृद्धि", "भविष्य",
    "परिवर्तन", "अवसर", "प्रेरणा", "उद्यम", "परिश्रम", "स्वास्थ्य", "समीक्षा", "विश्वास",
    "सहयोग", "उन्नति", "सम्पर्क", "भावना", "सन्दर्भ", "विचार", "प्रेरणादायक", "दृढता",
    "प्रयत्न", "निष्कर्ष", "उत्कृष्ट", "उपलब्धि", "परिकल्पना", "नविनता", "सहिष्णुता",
    "उदाहरण", "सन्तुलन", "आशा", "प्रयोजन", "समर्पित", "शान्ति", "आत्मविश्वास", "नयाँ",
    "सफल", "सकारात्मकता", "संघर्ष", "सहृदयता", "सहजता", "ज्ञानवर्धक", "सजिलो",
    "निश्चय", "संगठन", "समानता", "रचनात्मकता", "स्वतन्त्रता", "सहिष्णु", "जिम्मेवारी",
    "प्रतिबद्धता", "संयम", "उत्तम", "सकारात्मक_सोच"
]

# --- २. Lorem Ipsum जेनेरेटरको मुख्य तर्क (Main Logic) ---

def get_sentence(word_count=15):
    """
    तोकिएको संख्यामा शब्दहरू प्रयोग गरेर एउटा अनियमित नेपाली वाक्य बनाउँछ।
    """
    # शब्दहरूको सूचीबाट अनियमित रूपमा 'word_count' संख्या बराबर शब्दहरू छान्ने
    random_words = random.choices(nepali_words, k=word_count)
    
    # शब्दहरूलाई स्पेसले जोडेर अन्त्यमा पूर्णविराम (।) थप्ने
    return " ".join(random_words).replace("_", " ") + "।"

def get_nepali_lorem(paragraphs=3, sentences_per_paragraph=5, words_per_sentence=15):
    """
    नेपाली placeholder text (lorem ipsum) जेनेरेट गर्छ।
    """
    all_paragraphs = []
    for _ in range(paragraphs):
        paragraph_sentences = [get_sentence(words_per_sentence) for _ in range(sentences_per_paragraph)]
        all_paragraphs.append(" ".join(paragraph_sentences))
        
    return "\n\n".join(all_paragraphs)


# --- ३. Jupyter Notebook मा प्रयोगको लागि उदाहरण ---

print("--- वास्तविक नेपाली शब्दहरू मात्र भएको Lorem Ipsum ---")
final_lorem_text = get_nepali_lorem(
    paragraphs=2, 
    sentences_per_paragraph=4, 
    words_per_sentence=12
)
print(final_lorem_text)
```
