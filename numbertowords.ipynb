```python

# ========= 1. ERROR CLASSES =========
class Error(Exception):
    """Base class for custom errors."""
    pass

class UnsupportedLargeNumberError(Error):
    """Custom error for numbers exceeding the supported limit."""
    def __init__(self, number, max_number):
        self.number = number
        self.max_number = max_number

    def __str__(self):
        return f"Numbers larger than {self.max_number} not supported: {self.number}"

# ========= 2. NEPALI LANGUAGE DATA (As provided) =========
NUMBERS = {
  "0": "à¥¦", "1": "à¥§", "2": "à¥¨", "3": "à¥©", "4": "à¥ª",
  "5": "à¥«", "6": "à¥¬", "7": "à¥­", "8": "à¥®", "9": "à¥¯",
}

WORDS = {
  0: "à¤¸à¥à¤¨à¥à¤¨à¤¾", 1: "à¤à¤•", 2: "à¤¦à¥à¤ˆ", 3: "à¤¤à¥€à¤¨", 4: "à¤šà¤¾à¤°", 5: "à¤ªà¤¾à¤à¤š",
  6: "à¤›", 7: "à¤¸à¤¾à¤¤", 8: "à¤†à¤ ", 9: "à¤¨à¥Œ", 10: "à¤¦à¤¸", 11: "à¤à¤˜à¤¾à¤°",
  12: "à¤¬à¤¾à¤¹à¥à¤°", 13: "à¤¤à¥‡à¤¹à¥à¤°", 14: "à¤šà¥Œà¤§", 15: "à¤ªà¤¨à¥à¤§à¥à¤°", 16: "à¤¸à¥‹à¤¹à¥à¤°",
  17: "à¤¸à¤¤à¥à¤°", 18: "à¤…à¤ à¤¾à¤°", 19: "à¤‰à¤¨à¥à¤¨à¤¾à¤‡à¤¸", 20: "à¤¬à¥€à¤¸", 21: "à¤à¤•à¥à¤•à¤¾à¤‡à¤¸",
  22: "à¤¬à¤¾à¤‡à¤¸", 23: "à¤¤à¥‡à¤‡à¤¸", 24: "à¤šà¥Œà¤¬à¥€à¤¸", 25: "à¤ªà¤šà¥à¤šà¥€à¤¸", 26: "à¤›à¤¬à¥à¤¬à¥€à¤¸",
  27: "à¤¸à¤¤à¥à¤¤à¤¾à¤‡à¤¸", 28: "à¤…à¤ à¥à¤ à¤¾à¤‡à¤¸", 29: "à¤‰à¤¨à¤¨à¥à¤¤à¥€à¤¸", 30: "à¤¤à¥€à¤¸", 31: "à¤à¤•à¤¤à¥€à¤¸",
  32: "à¤¬à¤¤à¥à¤¤à¥€à¤¸", 33: "à¤¤à¥‡à¤¤à¥à¤¤à¥€à¤¸", 34: "à¤šà¥Œà¤à¤¤à¥€à¤¸", 35: "à¤ªà¥ˆà¤à¤¤à¥€à¤¸", 36: "à¤›à¤¤à¥à¤¤à¥€à¤¸",
  37: "à¤¸à¥ˆà¤à¤¤à¥€à¤¸", 38: "à¤…à¤ à¤¤à¥€à¤¸", 39: "à¤‰à¤¨à¤¨à¥à¤šà¤¾à¤²à¥€à¤¸", 40: "à¤šà¤¾à¤²à¥€à¤¸", 41: "à¤à¤•à¤šà¤¾à¤²à¥€à¤¸",
  42: "à¤¬à¤¯à¤¾à¤²à¥€à¤¸", 43: "à¤¤à¥à¤°à¤¿à¤šà¤¾à¤²à¥€à¤¸", 44: "à¤šà¤µà¤¾à¤²à¥€à¤¸", 45: "à¤ªà¥ˆà¤à¤¤à¤¾à¤²à¥€à¤¸", 46: "à¤›à¤¯à¤¾à¤²à¥€à¤¸",
  47: "à¤¸à¤¤à¤šà¤¾à¤²à¥€à¤¸", 48: "à¤…à¤ à¤šà¤¾à¤²à¥€à¤¸", 49: "à¤‰à¤¨à¤¨à¥à¤šà¤¾à¤¸", 50: "à¤ªà¤šà¤¾à¤¸", 51: "à¤à¤•à¤¾à¤‰à¤¨à¥à¤¨",
  52: "à¤¬à¤¾à¤‰à¤¨à¥à¤¨", 53: "à¤¤à¥à¤°à¤¿à¤ªà¤¨à¥à¤¨", 54: "à¤šà¤µà¤¨à¥à¤¨", 55: "à¤ªà¤šà¤ªà¤¨à¥à¤¨", 56: "à¤›à¤ªà¤¨à¥à¤¨",
  57: "à¤¸à¤¨à¥à¤¤à¤¾à¤‰à¤¨à¥à¤¨", 58: "à¤…à¤¨à¥à¤ à¤¾à¤‰à¤¨à¥à¤¨", 59: "à¤‰à¤¨à¤¨à¥à¤¸à¤¾à¤Ÿà¥à¤ à¥€", 60: "à¤¸à¤¾à¤ à¥€", 61: "à¤à¤•à¤¸à¤Ÿà¥à¤ à¥€",
  62: "à¤¬à¥ˆà¤¸à¤Ÿà¥à¤ à¥€", 63: "à¤¤à¥à¤°à¤¿à¤¸à¤Ÿà¥à¤ à¥€", 64: "à¤šà¥Œà¤à¤¸à¤Ÿà¥à¤ à¥€", 65: "à¤ªà¥ˆà¤à¤¸à¤Ÿà¥à¤ à¥€", 66: "à¤›à¥ˆà¤¸à¤Ÿà¥à¤ à¥€",
  67: "à¤¸à¤¤à¤¸à¤Ÿà¥à¤ à¥€", 68: "à¤…à¤ à¤¸à¤Ÿà¥à¤ à¥€", 69: "à¤‰à¤¨à¤¨à¥à¤¸à¤¤à¥à¤¤à¤°à¥€", 70: "à¤¸à¤¤à¥à¤¤à¤°à¥€", 71: "à¤à¤•à¤¹à¤¤à¥à¤¤à¤°",
  72: "à¤¬à¤¹à¤¤à¥à¤¤à¤°", 73: "à¤¤à¥à¤°à¤¿à¤¹à¤¤à¥à¤¤à¤°", 74: "à¤šà¥Œà¤¹à¤¤à¥à¤¤à¤°", 75: "à¤ªà¤šà¤¹à¤¤à¥à¤¤à¤°", 76: "à¤›à¤¯à¤¹à¤¤à¥à¤¤à¤°",
  77: "à¤¸à¤¤à¤¹à¤¤à¥à¤¤à¤°", 78: "à¤…à¤ à¤¹à¤¤à¥à¤¤à¤°", 79: "à¤‰à¤¨à¤¾à¤¸à¥€", 80: "à¤…à¤¸à¥€", 81: "à¤à¤•à¤¾à¤¸à¥€",
  82: "à¤¬à¤¯à¤¾à¤¸à¥€", 83: "à¤¤à¥à¤°à¤¿à¤¯à¤¾à¤¸à¥€", 84: "à¤šà¥Œà¤°à¤¾à¤¸à¥€", 85: "à¤ªà¤šà¤¾à¤¸à¥€", 86: "à¤›à¤¯à¤¾à¤¸à¥€",
  87: "à¤¸à¤¤à¤¾à¤¸à¥€", 88: "à¤…à¤ à¤¾à¤¸à¥€", 89: "à¤‰à¤¨à¤¾à¤¨à¥à¤¨à¤¬à¥à¤¬à¥‡", 90: "à¤¨à¤¬à¥à¤¬à¥‡", 91: "à¤à¤•à¤¾à¤¨à¤¬à¥à¤¬à¥‡",
  92: "à¤¬à¤¯à¤¾à¤¨à¤¬à¥à¤¬à¥‡", 93: "à¤¤à¥à¤°à¤¿à¤¯à¤¾à¤¨à¤¬à¥à¤¬à¥‡", 94: "à¤šà¥Œà¤°à¤¾à¤¨à¤¬à¥à¤¬à¥‡", 95: "à¤ªà¤¨à¥à¤šà¤¾à¤¨à¤¬à¥à¤¬à¥‡",
  96: "à¤›à¤¯à¤¾à¤¨à¤¬à¥à¤¬à¥‡", 97: "à¤¸à¤¨à¥à¤¤à¤¾à¤¨à¤¬à¥à¤¬à¥‡", 98: "à¤…à¤¨à¥à¤ à¤¾à¤¨à¤¬à¥à¤¬à¥‡", 99: "à¤‰à¤¨à¤¾à¤¨à¥à¤¸à¤¯",
}

AMOUNTS = {
  'saya': 'à¤¸à¤¯', 'hajaar': 'à¤¹à¤œà¤¾à¤°', 'lakh': 'à¤²à¤¾à¤–', 'crore': 'à¤•à¤°à¥‹à¤¡',
  'arab': 'à¤…à¤°à¥à¤¬', 'kharab': 'à¤–à¤°à¥à¤¬', 'neel': 'à¤¨à¥€à¤²', 'padma': 'à¤ªà¤¦à¥à¤®',
  'shankha': 'à¤¶à¤‚à¤–', None: ''
}

# ========= 3. PARSING LOGIC =========
MAX_NUMBER = 99_99_99_99_99_99_99_99_999
UNIT_AMOUNTS = [
  ('shankha', 1_00_00_00_00_00_00_00_000), ('padma', 1_00_00_00_00_00_00_000),
  ('neel', 1_00_00_00_00_00_000), ('kharab', 1_00_00_00_00_000),
  ('arab', 1_00_00_00_000), ('crore', 1_00_00_000), ('lakh', 1_00_000),
  ('hajaar', 1_000), ('saya', 100), (None, 1)
]

def get_chunks(number):
    """Breaks a number into chunks based on the Nepali numbering system."""
    if number > MAX_NUMBER:
        raise UnsupportedLargeNumberError(number, MAX_NUMBER)
    if number == 0:
        return [(0, None)]
    
    chunks = []
    remainder = number
    for name, amount in UNIT_AMOUNTS:
        if remainder < amount and amount > 1:
            continue
            
        quotient, remainder = divmod(remainder, amount)
        
        if quotient > 0:
            chunks.append((quotient, name))
            
    return chunks

# ========= 4. CORE CONVERSION FUNCTIONS =========
def to_nepali_words(number):
    """Converts an integer to Nepali words."""
    if not isinstance(number, int):
        raise TypeError("Input must be an integer.")

    if number < 100:
        return WORDS.get(number, "")

    chunks = get_chunks(number)
    
    word_parts = []
    for num, unit in chunks:
        # Recursively convert the number part (e.g., for 12 lakh, convert 12)
        num_word = to_nepali_words(num)
        unit_word = AMOUNTS.get(unit, "")
        word_parts.append(f"{num_word} {unit_word}".strip())
        
    return " ".join(word_parts)

def to_nepali_numerals(number):
    """Converts an integer to a string of Nepali numerals."""
    if not isinstance(number, int):
        raise TypeError("Input must be an integer.")
    
    return "".join([NUMBERS[digit] for digit in str(number)])

print("âœ… Nepali Number Converter functions are ready!")
```
