```python
# Cell 2: The Universal Converter Function

import re

def _legacy_to_unicode(text: str, source_font: str) -> str:
    """
    Converts text from a specified legacy font to Unicode using a rule-based approach.
    """
    if source_font not in ALL_FONT_RULES:
        return f"ERROR: Font '{source_font}' not supported. Available fonts: {list(ALL_FONT_RULES.keys())}"

    # Get the rules for the selected font
    rules = ALL_FONT_RULES[source_font]['rules']
    char_map = rules['character-map']
    post_rules = rules.get('post-rules', [])

    # 1. Apply the basic character map
    # Create a translation table for fast replacement
    mapped_text = text.translate(str.maketrans(char_map))
    
    # 2. Apply post-conversion regex rules for complex reordering and conjuncts
    for rule in post_rules:
        # The regex patterns in the JSON are strings, so we use them directly
        pattern, replacement = rule[0], rule[1]
        try:
            mapped_text = re.sub(pattern, replacement, mapped_text)
        except re.error as e:
            print(f"Warning: Invalid regex rule for {source_font}: '{pattern}' -> '{replacement}'. Error: {e}")
            
    return mapped_text

def _unicode_to_legacy(text: str, target_font: str) -> str:
    """
    Converts Unicode text back to a legacy font.
    Currently supports 'Preeti' as the target.
    """
    if target_font.lower() != 'preeti':
        return f"ERROR: Conversion to '{target_font}' is not supported. Only 'Preeti' is available as a target."

    # Sort the map keys by length (descending) to replace longer sequences first.
    sorted_keys = sorted(UNICODE_TO_PREETI_MAP.keys(), key=len, reverse=True)
    
    for uni_char in sorted_keys:
        text = text.replace(uni_char, UNICODE_TO_PREETI_MAP[uni_char])
        
    return text

def nepali_converter(text: str, source_format: str, target_format: str) -> str:
    """
    A complete converter for Nepali text.
    
    Args:
        text (str): The input text.
        source_format (str): The source font name (e.g., 'Preeti', 'Kantipur', 'Unicode').
        target_format (str): The target font name (e.g., 'Unicode', 'Preeti').
    """
    source = source_format.lower()
    target = target_format.lower()

    if target == 'unicode':
        # Capitalize the font name to match the keys in our rules dictionary
        return _legacy_to_unicode(text, source_font=source_format)
    
    elif source == 'unicode':
        return _unicode_to_legacy(text, target_font=target_format)
        
    else:
        # For legacy-to-legacy, first convert to Unicode, then to the target legacy font
        unicode_intermediate = _legacy_to_unicode(text, source_font=source_format)
        return _unicode_to_legacy(unicode_intermediate, target_font=target_format)

print("✅ `nepali_converter` function is ready for all supported fonts.")
```
